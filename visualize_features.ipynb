```python
import pandas as pd
import os
import numpy as np
```

```python
divorce_path = './code_book/'
feature_df = pd.read_csv(divorce_path + 'UCLA_data.csv')
feature_df=feature_df.replace(' ', np.nan, inplace=False).dropna(how='any')
feature_df.head()
```

```python
divorce_df = pd.read_csv(divorce_path + 'UCLA_Divorce_Data.csv')
divorce_df=divorce_df.replace(' ', np.nan, inplace=False).dropna(how='any')
divorce_df.head()
```

```python
join_frame = pd.merge(feature_df, divorce_df, on='csid')
join_frame.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
join_frame = join_frame.replace(' ', np.nan, inplace=False).dropna(how='any')
join_frame.head()
```

```python
plt.subplots(6,3,figsize=(15,15))

for i in range(1,19):
    plt.subplot(6,3,i)
    plt.title(join_frame.columns[i])
    sns.kdeplot(join_frame.loc[join_frame['Div']==0, join_frame.columns[i]], color='green', label=0)
    sns.kdeplot(join_frame.loc[join_frame['Div']==1, join_frame.columns[i]], color='blue', label=1) 
```
